## New data for $N=501$, with energy detunings

```python
points = np.linspace(-5,1.5,200)
points = 10**points

r_18 = np.load('calcs/energy_disorder-501-50nm-1.80--5.000-1.50-200.npy')
r_20 = np.load('calcs/energy_disorder-501-50nm-2.00--5.000-1.50-200.npy')
r_22 = np.load('calcs/energy_disorder-501-50nm-2.20--5.000-1.50-200.npy')
```

```python
f,ax = plt.subplots(figsize=(8,8))
ax.scatter(points, r_18, label=r'$E_M=1.8$')
ax.scatter(points, r_20, label=r'$E_M=2.0$')
ax.scatter(points, r_22, label=r'$E_M=2.2$')
ax.set_xscale('log')
ax.axhline(0.5307, c='k', linestyle='--', linewidth=1, label='GOE')
ax.axhline(0.3863, c='k', linestyle='--', linewidth=1, label='Poisson')
ax.set_xlabel(r'$\sigma(E_M)$ / $\Omega_R$')
ax.set_ylabel(r'$r$ index')
ax.legend()
ax.set_title(r'''$r$ index as a function of energetic disorder for the
$N=501$ photonic wire at different detunings''')
f.savefig('detuning.pdf')
```
