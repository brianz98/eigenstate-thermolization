```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.linspace(-5,-3,90)
points = 10**points

r_18 = np.load('calcs/r-everyother-1001-50nm-1.80--5.000--3.00-90.npy')
r_20 = np.load('calcs/r-everyother-1001-50nm-2.00--5.000--3.00-90.npy')
r_22 = np.load('calcs/r-everyother-1001-50nm-2.20--5.000--3.00-90.npy')

f,ax = plt.subplots(figsize=(10,4),sharex=True)
ax.scatter(points, r_18, marker='x',label=r'$E_M=1.8$')
ax.scatter(points, r_20, marker='x',label=r'$E_M=2.0$')
ax.scatter(points, r_22, marker='x',label=r'$E_M=2.2$')
ax.set_xscale('log')
ax.axhline(0.5307, c='k', linestyle='--', linewidth=1, label='GOE')
ax.axhline(0.3863, c='b', linestyle='--', linewidth=1, label='Poisson')
ax.legend()

ax.set_xlabel(r'$\sigma(E_M)$ / $\Omega_R$')
ax.set_ylabel(r'$r$ index')
    
f.suptitle(r'''$r$-index as a function of energetic disorder for the
    $N=1001$ photonic wire at different detunings, odd eigenstates only''')
f.savefig('every-other-1001.pdf')
```

```python
points = np.linspace(-5,1.5,200)
points = 10**points

r_18 = np.load('calcs/prod-251-50nm-1.80--5.000-1.50-200.npy')
r_20 = np.load('calcs/prod-251-50nm-2.00--5.000-1.50-200.npy')
r_22 = np.load('calcs/prod-251-50nm-2.20--5.000-1.50-200.npy')

f,ax = plt.subplots(3,1,figsize=(10,10),sharex=True)
for i in range(len(ax)):
    ax[i].scatter(points, r_18[:,i], marker='x',label=r'$E_M=1.8$')
    ax[i].scatter(points, r_20[:,i], marker='x',label=r'$E_M=2.0$')
    ax[i].scatter(points, r_22[:,i], marker='x',label=r'$E_M=2.2$')
    ax[i].set_xscale('log')
    if i==0:
        ax[i].axhline(0.5307, c='k', linestyle='--', linewidth=1, label='GOE')
        ax[i].axhline(0.3863, c='b', linestyle='--', linewidth=1, label='Poisson')
    ax[i].legend()

ax[-1].set_xlabel(r'$\sigma(E_M)$ / $\Omega_R$')
ax[0].set_ylabel(r'$r$ index')
ax[1].set_ylabel(r'Inverse participation ratio')
ax[2].set_ylabel(r'Shannon entropy')
    
f.suptitle(r'''Metrics as a function of energetic disorder for the
    $N=251$ photonic wire at different detunings''')
f.savefig('detuning-251.pdf')
```
