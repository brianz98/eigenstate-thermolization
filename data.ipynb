## Effect of ensemble size, a = 50nm, zero dipole and position disorder

```python
n501 = np.load('calcs/energy_disorder-0-2.5.npy')
n1001 = np.load('calcs/energy_disorder-1001-50nm-0-2.5.npy')
sigmas = np.linspace(0.001, 2.5, len(n501))
```

```python
f,ax = plt.subplots(figsize=(10,10))
ax.plot(sigmas, n501, marker='x', label=r'$N=501$')
ax.plot(sigmas, n1001, marker='^', label=r'$N=1001$')
ax.set_xlabel(r'$\sigma$ / $\Omega_r$')
ax.set_ylabel(r'$r$')
ax.legend()
```
